```python
from IPython.display import display, Markdown


array = [1200,1500,800,600,1000,400,900]
sumA = sum(array)
avg = sumA / len(array)
days = [x for x in array if x > 1000]
countDays = len(days)


display(Markdown(f"""
## Results

| Metric | Value |
|---|---:|
| Sum | {sumA} |
| Average | {avg:.2f} |
| Max | {max(array)} |
| Min | {min(array)} |
| Days > 1000 (count) | {countDays} |
"""))
```
