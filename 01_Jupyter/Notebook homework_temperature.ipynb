```python
from IPython.display import display, Markdown

temp = [2,5,4,6,7,3,1]
avg = sum(temp) / len(temp)
diff = max(temp) - min(temp)

display(Markdown(f"""
## Results

| Metric | Value |
|---|---:|
| Average | {avg} |
| Difference (max − min) | {diff} |
| Max | {max(temp)} |
| Min | {min(temp)} |
"""))
```
