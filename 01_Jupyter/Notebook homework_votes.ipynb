```python
from IPython.display import display, Markdown

votes = ["apple", "banana", "apple", "orange", "banana", "apple", "apple", "apple", "orange"]
dict_votes = {}
for vote in votes:
    if vote in dict_votes:
        dict_votes[vote] += 1
    else:
        dict_votes[vote] = 1

top_fruit = max(dict_votes, key=dict_votes.get)
print("The most popular fruit is:", top_fruit)

resStr = ""
for key, value in dict_votes.items():
    resStr += f"| {key} | {value} |\n"

display(Markdown(f"""
## Results

| Fruit | Votes |
|---|---:|
{resStr}
"""))

top_3_fruits = sorted(dict_votes, key=dict_votes.get, reverse=True)[:3]
print("The top 3 fruits are:", ", ".join(top_3_fruits))
```
